```python
# Code
import numpy as np
import pandas as pd
import seaborn as sns
from google.colab import drive
drive.mount('/content/drive')
```

```python
filepath='/content/drive/MyDrive/advance data wrangling assignment/processed.cleveland.data'
hd_df=pd.read_csv(filepath)
hd_df.columns=['age','sex','cp','trestbps','chol','fbs','resttecg','thalach','exang','oldpeak','slope','ca','thal','num']
hd_df.head()
```

```python
hd_df.info()
```

```python
hd_df.shape
```

```python
def age_bucket(age):
    if age >=30 and age <=49:
        return '30-49'
    elif age >=50 and age <=64:
        return '50-64'
    elif age >=65.0:
        return 'above 65'
    else:
        return 'Below 30'
hd_df['age_category']=hd_df['age'].apply(age_bucket)
hd_df.head()
```

```python
hd_df['max_heart_rate_category']=hd_df['thalach'].apply(lambda x:'>=150'if x>=150 else '<150')
```

```python
hd_df['max_heart_rate_category'].head(5)
```

```python
#aggregation operation
hd_df.groupby('sex').agg({'trestbps':['mean','median','max','min']})
```

```python
# NORMAL(<=200 mg/dL), borderline high(200-239 mg/dL),and high(>=240 mg/dL)
def chol_level(chol):
    if chol <=200:
        return 'normal (<=200 mg/dL)'
    elif chol >200 and chol <=239:
        return 'borderline high (200-239 mg/dL)'
    else:
        return 'high (>=240 mg/dL)'
hd_df['cholesterol_category']=hd_df['chol'].apply(chol_level)
hd_df.head()
```

```python
hd_df.groupby('age_category').agg({'chol':['mean','std']})
```

```python
hd_df.groupby(['sex','cp']).agg({'chol':['mean','median','max','min']})
```

```python
hd_df.head(5)
```

```python
 # No ST depression, slight ST depression (<=1 mm) and significant ST depresssion (>1 mm)
def deep_level(d):
    if d >1:
        return 'significsnt depression (>1 mm)'
    elif d <=1:
        return 'slight depression (<=1 mm)'
    else:
        return 'no ST depression'
hd_df['ST_depression_category']=hd_df['oldpeak'].apply(deep_level)
hd_df.head()
```

```python
hd_df.groupby(['sex','age_category']).agg({'oldpeak':['mean','median','std']})
```

### **Honey Production in the USA (1998-2012)**

Suppose you have been provided with a dataset of Honey Production in the USA from the years 1998-2012. The dataset contains information about the state, number of colonies, yield per colony, total production, stocks, price per pound, production value, and year. You want to analyze this data to gain insights into the honey production in the USA.

Using the provided dataset, perform the following tasks:

1. Split the dataset into two separate datasets, one for odd years and one for even years.
2. Group the data by state and calculate the total production for each state for each year.
3. Calculate the average yield per colony for each state over the entire dataset.
4. Left join the odd year dataset with the even year dataset to create a new dataset containing the total production for each state for both odd and even years.
5. Right join the odd year dataset with the even year dataset to create a new dataset containing the total production for each state for both odd and even years.

```python
# Write your code here
import numpy as np
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')
```

```python
filepath='/content/drive/MyDrive/advance data wrangling assignment/honeyproduction.csv'
df=pd.read_csv(filepath)
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df_even=df.loc[df.loc[:,'year']%2 == 0,:]
df_odd=df.loc[df.loc[:,'year']%2 != 0,:]
print(df_even.shape,df_odd.shape)
```

```python
df_even.head()
```

```python
df_odd.head()
```

```python
df.groupby(['state','year'])['totalprod'].sum()
```

```python
df.groupby(['state'])['yieldpercol'].mean()
```

```python
e=df_even.groupby('state').agg(even_total_prod=('totalprod','sum')).reset_index()
```

```python
o=df_odd.groupby('state').agg(odd_total_prod=('totalprod','sum')).reset_index()
e.head()
```

```python
o.head()
```

```python
pd.merge(e,o, how='left', on='state').head()
```

```python
pd.merge(e,o, how='right', on='state').head()
```
